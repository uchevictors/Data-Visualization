```python
#Ploting a graph
```

```python
albany = df.copy()[df['region'] == "Albany"]
df["Date"] = pd.to_datetime(df["Date"])
albany = albany.set_index("Date")
albany.sort_index(inplace=True)
albany['price25ma'] = albany['AveragePrice'].rolling(25).mean()
albany.head()
```

```python
albany['price25ma'] = albany['AveragePrice'].rolling(25).mean().plot(xlabel
='Date', ylabel='Average Price', title='Average Prices Of Avocado In Albany ($)')
```

```python
albany['price25ma'] = albany['AveragePrice'].rolling(25).max().plot(xlabel='Date', ylabel='Average Price', title='Average Prices Of Avocado In Atlanta - Max. Price = $2.15')
```

```python
albany['price25ma'] = albany['AveragePrice'].rolling(25).min().plot(xlabel='Date', ylabel='Average Price', title='Average Prices Of Avocado In Atlanta - Min. Price = $0.85')
```
